```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Day vs Temp")
plt.plot(x, y, color = 'green', linewidth=2, linestyle='dashed')
```

```python
plt.plot(x, y, 'd-.r', markersize=5, alpha = 0.75)
```

```python
days = [1, 2, 3, 4, 5, 6,7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Temp vs Day")
plt.plot(days, max_t, label= 'Max')
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Average")
plt.legend(loc="upper right", shadow = True, fontsize='small')
plt.grid()
```

## Bar plots

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit = [40, 5, 34, 12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.ylabel('Amount (bln)')
plt.title('US Tech stocks')
plt.bar(xpos - 0.2, revenue, label='Revenue', width = 0.4)
plt.bar(xpos + 0.2, profit, label='Profit', width = 0.4)
plt.legend()
```

```python

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html

plt.yticks(xpos, company)
plt.title('US Tech stocks')
plt.barh(xpos - 0.2, revenue, label='Revenue', height=0.4)
plt.barh(xpos + 0.2, profit, label='Profit', height=0.4)
plt.legend()
```

## Histograms

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93,
                 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84,
                    69, 89, 79, 120, 112, 100]


plt.xlabel('sugar range')
plt.ylabel('Total number of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150],
        rwidth = 0.95, color=['green', 'orange'], label=['men', 'women'])
plt.legend()
plt.show()
```

```python
plt.xlabel('sugar range')
plt.ylabel('Total number of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150],
        rwidth = 0.95, color=['green', 'orange'], label=['men', 'women'],
        orientation= 'horizontal')
plt.legend()
plt.show()
```

```python
plt.xlabel('sugar range')
plt.ylabel('Total number of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150],
        rwidth = 0.95, color=['green', 'orange'], label=['men', 'women'],
        orientation= 'horizontal')
plt.legend()
plt.savefig("C:/Users/Public/lmaaya/projects/python/JupiterNoteBook/graphs/horizontal_hist.png",
           bbox_inches = 'tight')
```

## Pie chart

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", 
              "Car", "Other utilities"]
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels= exp_labels, radius=1.5, autopct='%0.1f%%',
       shadow=True, explode = [0,0.2,0,0,0.3],
       startangle= 180)
plt.show()
```
